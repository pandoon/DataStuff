```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\thoma\Desktop\Projects\Movie Review\Play Test - Movie History.csv'
dataset = pd.read_csv(file_path)
```

```python
dataset = dataset.drop(columns='Date Watched')
```

```python
dataset
```

```python
def convert_rating(text_rating):
    try:
        text_rating = str(text_rating)
        parts = text_rating.split("/")
        numerator = float(parts[0])
        denominator = float(parts[1])
        numerical_rating = numerator/denominator
        return numerical_rating
    
    except (ValueError, IndexError, ZeroDivisionError) as e:
        print(f"Error converting rating '{text_rating}': {e}")
        return None
```

```python
dataset['Jesus NScore'] = dataset["Jack's Score"].apply(convert_rating)
```

```python
dataset['Abhraham NScore'] = dataset["Aaron's Score"].apply(convert_rating)
```

```python
dataset["Thomas' NScore"] = dataset["Thomas' Score"].apply(convert_rating)
```

```python
dataset = dataset.drop(columns="Jack's Score")
```

```python
dataset = dataset.drop(columns="Aaron's Score")
```

```python
dataset = dataset.drop(columns="Thomas' Score")
```

```python
dataset
```

```python
mean_genre_rating = dataset.groupby('Genre').mean()
```

```python
mean_genre_rating = mean_genre_rating.dropna()
```

```python
mean_genre_rating.plot.bar(rot=0, figsize=(18, 6))
plt.title('Average Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.legend(title='Reviewer', loc='upper right')
plt.show()
```
